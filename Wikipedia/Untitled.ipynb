## Read file

```python
f = open('wikipedia.txt.rtf', 'r')
data = f.read()
data = data.split('\n')
print(data)
```

```python
## words = []
allwords = []
keywords = []
titles = []
keyword_list = []

#titles += [line for line in data[9:] if '==' is line[:2]]
for lines in data[10:20]:
    index = 0
    if lines[:2] == '==':
        title = lines.translate({ord(c):'' for c in "="})
        #finalt = title.translate({ord(c):'' for c in "['\\96]" })
        titles.append(title[:-1])
        continue
    lines = lines.split(' ')
    words = list(filter(lambda x: 'asdhagsdgahsd' not in x, lines))

    for word in words:
        
        if word[:2] == '[[':
            
            index2 = index
            while ']]' or '|' not in words[index2]:
                index2 +=1
            if '|' in words[index2]:
                keyword = words[index2].split('|')[-1][:-2]
            else:
                keyword = ' '.join(words[index:index2+1])[2:-2]
            keyword_list.append(keyword)
        index += 1
    allwords.append(words)
print(keyword_list)
#print(allwords)
#print(titles)
#print(allwords[0])
```
